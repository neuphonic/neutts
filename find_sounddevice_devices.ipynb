```python
import sounddevice as sd

print("=======================================")
print(" ALL AVAILABLE AUDIO DEVICES")
print("=======================================")
# This prints the raw list. Look for the '>' symbol, 
# which indicates your default input device.
print(sd.query_devices())

print("\n=======================================")
print(" YOUR MAC'S DEFAULT MICROPHONE")
print("=======================================")
# sd.default.device returns a tuple: (default_input_index, default_output_index)
default_input_index = sd.default.device[0]
device_info = sd.query_devices(default_input_index)

print(f"Device Index : {default_input_index}")
print(f"Device Name  : {device_info['name']}")
print(f"Sample Rate  : {device_info['default_samplerate']} Hz")
print(f"Max Channels : {device_info['max_input_channels']}")
print("=======================================\n")
```
